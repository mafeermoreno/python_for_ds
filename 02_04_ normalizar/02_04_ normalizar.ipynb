# Normalize the columns of a dataframe

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('datos.csv', delimiter=";")
```

```python
print(df.head(20))
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
print(scaled_data)
```

Then, we need to turn the given array into a new dataframe with scaled 'Altura' and 'Peso' columns

```python
scaled_df = pd.DataFrame(scaled_data, columns=['Altura', 'Peso'])
print(scaled_df.head(20))
```
